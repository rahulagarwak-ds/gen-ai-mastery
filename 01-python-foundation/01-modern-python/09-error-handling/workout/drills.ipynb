### Drill B2: Custom Exception Class ðŸ”´

```python
# Create custom exceptions
class APIError(Exception):
    """Base exception for API errors."""
    pass

class AuthenticationError(APIError):
    """Authentication failed."""
    pass

class RateLimitError(APIError):
    """Rate limit exceeded."""
    def __init__(self, retry_after: int):
        super().__init__(f"Rate limited. Retry after {retry_after}s")
        self.retry_after = retry_after

# Test
try:
    raise RateLimitError(60)
except RateLimitError as e:
    print(f"Error: {e}")
    print(f"Retry in: {e.retry_after}s")
```

### Drill B3: Exception Chaining ðŸ”´
**Task:** Chain exceptions to preserve original cause

```python
import json

class ConfigError(Exception):
    pass

def load_config(json_string):
    """Parse JSON config, chain exceptions if it fails."""
    try:
        return json.loads(json_string)
    except json.JSONDecodeError as e:
        # Chain to ConfigError (use 'from e')
        pass

# Test
try:
    load_config("{invalid json}")
except ConfigError as e:
    print(f"ConfigError: {e}")
    print(f"Caused by: {e.__cause__}")
```
