### Drill D2: groupby ðŸ”´
**Task:** Group consecutive items

```python
# Note: groupby requires sorted data!
data = [
    {"type": "fruit", "name": "apple"},
    {"type": "fruit", "name": "banana"},
    {"type": "veggie", "name": "carrot"},
    {"type": "veggie", "name": "broccoli"},
    {"type": "fruit", "name": "cherry"},
]

# Sort first, then group by type
sorted_data = sorted(data, key=lambda x: x["type"])
for type_name, group in groupby(sorted_data, key=lambda x: x["type"]):
    names = [item["name"] for item in group]
    print(f"{type_name}: {names}")

# Expected:
# fruit: ['apple', 'banana', 'cherry']
# veggie: ['carrot', 'broccoli']
```

### Drill D3: Combinations and Permutations ðŸŸ¡
**Task:** Generate combinations and permutations

```python
teams = ["A", "B", "C", "D"]

# All possible matchups (combinations of 2)
matchups = list(combinations(teams, 2))
print(f"Matchups: {matchups}")
# [('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'C'), ('B', 'D'), ('C', 'D')]

# Ordered pairs (permutations of 2) - order matters
ordered_pairs = list(permutations(teams, 2))
print(f"Ordered pairs: {len(ordered_pairs)}")  # 12
```
