### Drill C4: LEGB Rule üî¥
**Task:** Predict the output of each print statement.

**Expected Output:**
```
Inner: local
Outer: enclosing
Global: global
```

````python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print(f"Inner: {x}")

    inner()
    print(f"Outer: {x}")

outer()
print(f"Global: {x}")
````
